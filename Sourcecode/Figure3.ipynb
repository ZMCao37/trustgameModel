```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Generate random data for scatter plot
np.random.seed(878)
x = np.random.uniform(low=-1, high=1, size=4)
y = np.random.uniform(low=-1, high=1, size=4)
labels = ['Subject1', 'Subject2', 'Subject3', 'Subject4']

# Plot the scatter plot
plt.scatter(x, y)

# Annotate each point with its label and an arrow pointing to it
for i, label in enumerate(labels):
    plt.annotate(label, xy=(x[i], y[i]), xytext=(x[i] + 0.3*np.sign(x[i]), y[i] + 0.3*np.sign(y[i])),
                 arrowprops=dict(facecolor='black', shrink=0.05))

# Set the x and y axis limits and labels
plt.xlim(-1, 1)
plt.ylim(-1, 1)
plt.xlabel('wa')
plt.ylabel('wt')

# Uncomment the next line if you want to add a legend
# plt.legend(labels)

# Display the scatter plot
plt.show()
```

```python
# Generate x values for the sigmoid functions
x = np.linspace(-1, 8, 100)

# Generate random parameters for the sigmoid functions
np.random.seed(2)
params = np.random.uniform(low=1, high=6, size=(4, 2))

# Plot four sigmoid functions, each representing the probability of investment for a subject
for i in range(4):
    y = 1 / (1 + np.exp(-params[i][0] * (x - params[i][1])))
    plt.plot(x, y, label=f'Subject{i+1}')

# Set the x and y axis labels and a title
plt.xlabel('Evaluation')
plt.ylabel('Probablity of investment')

# Add a legend to the plot
plt.legend()

# Display the sigmoid functions
plt.show()
```

```python
# Generate x values for the lines
x = np.linspace(-3, 10, 100)

# Generate random parameters for the lines
np.random.seed(267890)
params = np.random.uniform(size=(4, 2))
params[:, 0] *= 1500 / 10  # Map the slopes to the range 0-1500
params[:, 1] *= 1500      # Map the intercepts to the range 0-1500

# Plot four lines, each representing the monetary score of a subject as a function of evaluation
for i in range(4):
    y = params[i][0] * x + params[i][1]
    if i < 2:
        # For the first two subjects, plot the first 20% of the x values as dashed lines
        plt.plot(x[:20], y[:20], '--', label=f'subject{i+1}')
        # Plot the remaining x values as solid lines
        plt.plot(x[20:], y[20:], '-', color=plt.gca().lines[-1].get_color())
    else:
        # For the last two subjects, plot the entire line as solid
        plt.plot(x[:20], y[:20], '--')
        plt.plot(x[20:], y[20:], '-', label=f'subject{i+1}')

# Set the x and y axis labels
plt.xlabel('Evaluation')
plt.ylabel('Monetary score')

# Uncomment the next line if you want to add a title to the plot
# plt.title('Random lines')

# Add a legend to the plot
plt.legend()

# Display the plot
plt.show()
```
